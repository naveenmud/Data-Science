```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located

file = "Resources/crime_incident_data2017.csv"

# Import the data into a Pandas DataFrame

df = pd.read_csv(file)
df.head()
```

```python
# look for missing values
df.count()
```

```python
# drop null rows
df = df.dropna(how='any')
```

```python
# verify counts
df.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
df['Offense Type'].value_counts()
```

```python
# Combine similar offenses

df['Offense Type'] = df['Offense Type'].replace(
    {'Purse-Snatching': 'Pocket-Picking'})

df['Offense Type'] = df['Offense Type'].replace(
    {'Commercial Sex Acts': 'Prostitution', 'Assisting or Promoting Prostitution': 'Prostitution'})

df['Offense Type'] = df['Offense Type'].replace(
    {'Theft From Motor Vehicle': 'Motor Vehicle Theft', 'Theft of Motor Vehicle Parts or Accessories': 'Motor Vehicle Theft'})
```

```python
# Check to see if you comnbined similar offenses correctly in "Offense Type".

df['Offense Type'].value_counts()
```

```python
# Create a new DataFrame that looks into a specific neighborhood

new_df = df.loc[df["Neighborhood"] == "Montavilla", :]
new_df.head()
```
