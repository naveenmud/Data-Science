```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_axis = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
x_axis = np.arange(1, 13, 1)
```

```python
celsius = [(y - 32) * 0.56 for y in y_axis]
```

```python
plt.plot(x_axis,y_axis)
plt.plot(x_axis,celsius)
plt.show()
```

```python

y_axis = plt.plot(x_axis, y_axis, marker ='+', color='blue', label="Fahrenheit")

celsius = plt.plot(x_axis, celsius, marker='^', color='red', label="Celsius")
```

```python
plt.legend(loc="upper right")
```

```python
plt.xlabel("Months")
plt.ylabel("Degrees")
```
