```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis = np.arange(0, 130, 10)
danger = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.plot(x_axis,danger)
plt.plot(x_axis,rail)
plt.show()
```

```python
plt.plot(x_axis, danger, linewidth=1, color="red", label = "Danger Drop")
plt.plot(x_axis, rail, linewidth=1, color="green", label = "Rail Gun")
```

```python
plt.legend(loc="upper right")
```

```python
plt.grid()
```
