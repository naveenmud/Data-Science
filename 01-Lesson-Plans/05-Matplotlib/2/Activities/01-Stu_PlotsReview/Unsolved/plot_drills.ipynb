```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
plt.bar(gyms, members, color="b", align="center")
tick_locations = [value for value in gyms]
plt.xticks(tick_locations, gyms)
plt.xlim(-0.75, len(gyms)-0.25)
plt.ylim(0, max(members)+10)
plt.title("Gym Size")
plt.xlabel("Gym Name")
plt.ylabel("Number of Members")
plt.show()
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)

plt.plot(x_axis, sin)
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)

plt.pie(members, explode=explode, labels = gyms, colors=colors,
        autopct="%1.0f%%", startangle=140)
plt.show()
```

```python
# DATASET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
    
plt.scatter(x_axis, times, marker="o", facecolors="red", edgecolors="black")
```
